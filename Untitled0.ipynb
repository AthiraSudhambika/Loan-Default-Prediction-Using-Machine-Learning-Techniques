```python
from sklearn.impute import SimpleImputer

numerical_cols = df.select_dtypes(include=['float64', 'int64']).columns
categorical_cols = df.select_dtypes(include=['object']).columns

imputer_num = SimpleImputer(strategy='median')
imputer_cat = SimpleImputer(strategy='most_frequent')

df[numerical_cols] = imputer_num.fit_transform(df[numerical_cols])
df[categorical_cols] = imputer_cat.fit_transform(df[categorical_cols])
```

```python
df = pd.get_dummies(df, columns=categorical_cols, drop_first=True)
```

```python
import pandas as pd
import matplotlib.pyplot as plt
from google.colab import files

# Step 1: Upload the dataset
uploaded = files.upload()

# Step 2: Load the uploaded file
file_name = list(uploaded.keys())[0]  # Get the name of the uploaded file
df = pd.read_csv(file_name)

# Display the first few rows to understand the data
#print("First few rows of the dataset:")
#print(df.head())

# Step 3: Preprocess the data
# Check unique values in 'loan_status' to understand its current state
print("\nUnique values in 'loan_status' before processing:")
print(df['loan_status'].unique())

# Define the mapping dictionary for the target variable 'loan_status'
loan_status_map = {
    'Fully Paid': 0,
    'Charged Off': 1,
    'Default': 1,
    'Late (31-120 days)': 1
}

# Replace the categorical values in 'loan_status' with numerical values
df['loan_status'] = df['loan_status'].map(loan_status_map)

# Drop rows where 'loan_status' is NaN (those that were not mapped)
df = df.dropna(subset=['loan_status'])

# Ensure 'loan_status' is of type integer
df['loan_status'] = df['loan_status'].astype(int)

# Verify the replacement
print("\nUnique values in 'loan_status' after processing:")
print(df['loan_status'].unique())

# Create a dictionary to map back from numeric to descriptive labels for plotting
reverse_loan_status_map = {v: k for k, v in loan_status_map.items()}

# Step 4: Plotting the bar chart
plt.figure(figsize=(6, 4))
value_counts = df['loan_status'].value_counts()
value_counts.plot(kind='bar', color=['skyblue', 'salmon'])

# Add labels and title
plt.title('Distribution of Loan Statuses')
plt.xlabel('Loan Status')
plt.ylabel('Count')
plt.xticks(ticks=range(len(value_counts)), labels=[reverse_loan_status_map[i] for i in value_counts.index], rotation=0)

# Display the plot
plt.show()

```
