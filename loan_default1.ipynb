```python
import pandas as pd
from sklearn.impute import SimpleImputer
from google.colab import files

# Upload the dataset
uploaded = files.upload()

# Assuming the uploaded file is named 'your_dataset.csv'
df = pd.read_csv('loans_full_schema.csv')

# Impute missing values in numeric columns with the median
numeric_imputer = SimpleImputer(strategy='median')
numeric_columns = df.select_dtypes(include=['number']).columns
df[numeric_columns] = numeric_imputer.fit_transform(df[numeric_columns])

# Impute missing values in categorical columns with the most frequent value
categorical_imputer = SimpleImputer(strategy='most_frequent')
categorical_columns = df.select_dtypes(include=['object']).columns
df[categorical_columns] = categorical_imputer.fit_transform(df[categorical_columns])

# Check remaining null values
print(df.isnull().sum())

# Save the cleaned dataset to a new CSV file
df.to_csv('cleaned_dataset.csv', index=False)
print("Cleaned dataset saved as 'cleaned_dataset.csv'")
```

```python
import pandas as pd
from sklearn.impute import KNNImputer
from google.colab import files

# Upload the dataset
uploaded = files.upload()

# Assuming the uploaded file is named 'your_dataset.csv'
df = pd.read_csv('loans_full_schema.csv')

# Initialize KNN Imputer
knn_imputer = KNNImputer(n_neighbors=5)

# Apply KNN imputation to numeric columns
numeric_columns = df.select_dtypes(include=['number']).columns
df[numeric_columns] = knn_imputer.fit_transform(df[numeric_columns])

# For categorical columns, you can still use the most frequent value imputer
from sklearn.impute import SimpleImputer
categorical_imputer = SimpleImputer(strategy='most_frequent')
categorical_columns = df.select_dtypes(include=['object']).columns
df[categorical_columns] = categorical_imputer.fit_transform(df[categorical_columns])

# Check remaining null values
print(df.isnull().sum())

# Save the cleaned dataset to a new CSV file
df.to_csv('cleaned_dataset_knn.csv', index=False)
print("Cleaned dataset saved as 'cleaned_dataset_knn.csv'")

```

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from google.colab import files
import io

# Step 1: Upload the file
uploaded = files.upload()

# Assuming only one file is uploaded, get the file name
file_name = list(uploaded.keys())[0]

# Load the dataset
try:
    df = pd.read_csv(io.BytesIO(uploaded[file_name]))
    print("Data loaded successfully.")
except FileNotFoundError:
    raise ValueError(f"The file was not found.")
except pd.errors.EmptyDataError:
    raise ValueError("The file is empty. Please check the file content.")
except Exception as e:
    raise ValueError(f"An error occurred while loading the data: {e}")

# Check if DataFrame is empty
if df.empty:
    raise ValueError("The DataFrame is empty. Please check the data loading process.")

# Print the first few rows of the DataFrame to verify data
print("First few rows of the DataFrame:")
print(df.head())

# Check unique values in 'loan_status' before replacement
print("Unique values in 'loan_status' before replacement:")
unique_values_before = df['loan_status'].unique()
print(unique_values_before)

# Example replacement logic (adjust according to your needs)
# Replace 'loan_status' with numerical values
replacement_dict = {
    'Fully Paid': 1,
    'Charged Off': 0,
    'Current': 2,
    'In Grace Period': 3,
    'Late (31-120 days)': 4,
    'Late (16-30 days)': 5
}
df['loan_status'] = df['loan_status'].replace(replacement_dict)

# Check unique values in 'loan_status' after replacement
print("Unique values in 'loan_status' after replacement:")
unique_values_after = df['loan_status'].unique()
print(unique_values_after)

# Ensure that 'loan_status' column is now numerical
try:
    df['loan_status'] = df['loan_status'].astype(int)
    print("Conversion to int successful.")
except ValueError as e:
    print(f"Error converting 'loan_status' to int: {e}")
    print("There are still non-numeric values in 'loan_status'. Please check the replacement logic.")
    raise

# Check if the DataFrame is still empty after replacements
if df.empty:
    raise ValueError("The DataFrame is empty after replacements. Please check the replacement logic.")

# Split the dataset into train and test sets with stratification on 'loan_status'
train, test = train_test_split(df, test_size=0.2, random_state=1, stratify=df['loan_status'])

# Output the shapes of the resulting splits
print("\nTraining set shape:", train.shape)
print("Test set shape:", test.shape)
```
